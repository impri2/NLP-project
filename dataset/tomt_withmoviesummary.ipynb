```python
%cd /content/drive/MyDrive/dataset
import sys
import gc
sys.path.insert(0,'/content/drive/MyDrive/dataset/ColBERT')
sys.path.insert(0,'/content/drive/MyDrive/dataset')
import torch
import data
import ColBERT.colbert.modeling.weighted_colbert as colbert
from colbert.modeling.tokenization.query_tokenization import QueryTokenizer
from  colbert.modeling.tokenization.doc_tokenization import DocTokenizer
from transformers import BertConfig
from torch.optim import Adam
import pickle
documentData = dict()
with open('document.pickle','rb') as f:
 documentData = pickle.load(f)
dataset = data.TomtDataset(documentData)
dataloader = torch.utils.data.DataLoader(dataset,batch_size=4,shuffle=True)
query_maxlen = 512
doc_maxlen = 512
mask_punctuation = True
config = BertConfig()
model = colbert.WeightedColBERT(config,query_maxlen,doc_maxlen,mask_punctuation).to('cuda:0')
model.train()
optimizer = Adam(model.parameters(),lr = 2e-5)
queryTokenizer = QueryTokenizer(query_maxlen)
docTokenizer=DocTokenizer(doc_maxlen)
lossFunction = torch.nn.CrossEntropyLoss()
num_epoch = 5
for epoch in range(num_epoch):
 for batch,d in enumerate(dataloader):

  optimizer.zero_grad()
  queries = d['query']
  documents=d['document']


  negatives=d['negative']


  scores1 = model.forwardInBatchAndNegatives(queryTokenizer.tensorize(queries),docTokenizer.tensorize(documents[0]),docTokenizer.tensorize(negatives[0]))
  scores2 = model.forwardInBatchAndNegatives(queryTokenizer.tensorize(queries),docTokenizer.tensorize(documents[1]),docTokenizer.tensorize(negatives[1]))

  loss = lossFunction(scores1+scores2,torch.tensor([0,1,2,3]).to('cuda:0'))
  loss.backward()
  optimizer.step()
  if batch% 100 == 0:
    print(str(batch) + ' ' +str(loss))
 torch.save(model.state_dict(),'moredata_epoch'+str(epoch+1)+'.pt')
```

```python
%cd /content/drive/MyDrive/dataset
import sys
import gc
from transformers import BertConfig
sys.path.insert(0,'/content/drive/MyDrive/dataset/ColBERT')
sys.path.insert(0,'/content/drive/MyDrive/dataset')
from inference import *
import ColBERT.colbert.modeling.weighted_colbert as colbert
import pickle
documentData = dict()
with open('document.pickle','rb') as f:
 documentData = pickle.load(f)
query_maxlen = 512
doc_maxlen = 512
mask_punctuation = True
config = BertConfig()
model = colbert.WeightedColBERT(config,query_maxlen,doc_maxlen,mask_punctuation).to('cuda:0')
model.load_state_dict(torch.load('moredata_epoch5.pt'))
embeddings = get_document_embeddings(model,documentData)
```

```python
import inference
import data as d
validation = open('/content/drive/MyDrive/dataset/Movies/splits/validation/qrels.txt')
model.eval()
modelInference = inference.ModelInference(model)
tokenizer = QueryTokenizer(query_maxlen)
queries = d.load_queries()
recall10 = 0
i = 0
for line in validation.readlines():
 try:
  query = queries[line.split()[0]]
  document = line.split()[2]
  docEmbed = embeddings[document]
  Q = modelInference.query(*tokenizer.tensorize([query]))

  score = model.score(Q,embeddings[document][0])+model.score(Q,embeddings[document][1])
  rank = 1
  for doc in embeddings:
    if doc == document:continue
    if model.score(Q,embeddings[doc][0])+model.score(Q,embeddings[doc][1])>score:rank+=1
  if rank <=10:recall10 += 1
  i += 1
  print(rank)
  #if i == 100: break
 except:
  print('no data')
  continue

print(recall10)




```
